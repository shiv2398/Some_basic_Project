```python
def year_extract(text):
    year_start=text.find('(')
    year_end=text.find(')')
    return text[year_start+1:year_end]
def title_extract(text):
    cl_text=text.split('(')[0].strip()
    return cl_text
import datetime
def year_change(year_str):
    current_year = datetime.datetime.now().year  # Get the current year
    
    try:
        year = int(year_str)  # Convert the year string to an integer
        total_years = current_year - year  # Calculate the total number of years
        return total_years
    except ValueError:
        return -1  # Default value for invalid year strings
```

```python
from nltk.tokenize import word_tokenize

mdf['movie_year']=mdf['title'].apply(year_extract)
mdf['movie_title']=mdf['title'].apply(title_extract)
mdf['movie_title']=mdf['movie_title'].apply(word_tokenize)
mdf['movie_year']=mdf['movie_year'].apply(year_change)
mdf.drop('title',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X=mdf.drop('rating',axis=1)
y=mdf['rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.2, random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
#scaling the features 
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled.shape,y_train.shape,X_test_scaled.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

def train_and_test(x_train, y_train, x_test, y_test):
    # Define hyperparameters for each model
    regression_models = {
        'Linear Regression': LinearRegression(),
        'Ridge Regression': Ridge(alpha=1.0),
        'Lasso Regression': Lasso(alpha=1.0),
        'ElasticNet Regression': ElasticNet(alpha=1.0, l1_ratio=0.5),
        'Decision Tree Regression': DecisionTreeRegressor(max_depth=5),
        'Random Forest Regression': RandomForestRegressor(n_estimators=100),
        'Gradient Boosting Regression': GradientBoostingRegressor(n_estimators=100),
        'Support Vector Regression': SVR(C=1.0, epsilon=0.1),
        'K-Nearest Neighbors Regression': KNeighborsRegressor(n_neighbors=5),
        #'Neural Network Regression': MLPRegressor(hidden_layer_sizes=(100,))
    }

    mse_dict = {}
    trained_models = {}

    for name, model in regression_models.items():
        print(f'Training : {name}')
        model.fit(x_train, y_train)
        trained_models[name] = model

        # Reshape x_test to a 2D array
        #x_test_reshaped = x_test.reshape(-1, 1)

        y_pred = model.predict(x_test)

        # Calculate RMSE
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mse_dict[name] = rmse

    return mse_dict, trained_models

```

```python
mse_,trained_models=train_and_test(X_train_scaled,y_train,X_test_scaled,y_test)
```

```python
rmse_df = pd.DataFrame.from_dict(mse_, orient='index', columns=['RMSE'])
rmse_df.reset_index(inplace=True)
rmse_df.rename(columns={'index': 'Model'}, inplace=True)
rmse_df = rmse_df.sort_values(by='RMSE')  # Sort the DataFrame by RMSE values in ascending order
lowest_rmse_model = rmse_df.iloc[0]  # Select the model with the lowest RMSE value

# Display the DataFrame sorted by RMSE values
sorted_rmse_df = rmse_df.style.highlight_min(subset=['RMSE'], color='lightgreen', axis=0)
sorted_rmse_df
```

```python
best_model_key=rmse_df.loc[rmse_df['RMSE'].idxmin()]
best_model=trained_models[best_model_key[0]]
best_model
```

```python
#saving the model and the scaler

import pickle

with open(best_model_key[0]+'.pkl','wb') as f:
    pickle.dump(best_model,f)
with open('scaler.pkl','wb') as f:
    pickle.dump(scaler,f)
```
